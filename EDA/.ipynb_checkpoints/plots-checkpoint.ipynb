```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

## packed red cell

```python
packed_red_cells = pd.read_csv('Q2.csv')
packed_red_cells
```

```python
packed_red_cells.columns
```

```python
sns.distplot(packed_red_cells['totalamount'], kde=False)
set(packed_red_cells['totalamount'])
```

```python
packed_red_cells = pd.read_csv('Q2.csv')
print(packed_red_cells['totalamount'].describe())
packed_red_cells_count_df = packed_red_cells.groupby('totalamount').count().sort_values('hadm_id')

plt.figure(figsize=(70,7))
sns.barplot(x=packed_red_cells_count_df.index, y=packed_red_cells_count_df.hadm_id)
plt.ylabel('Prescription Category Frequency', fontsize=20)
plt.xlabel('Presciption Category', fontsize=20)
plt.show()
```

## Last hemoglobin before blood transfusion male compare to female

```python
last_Hb = pd.read_csv('Q3.csv')
last_Hb 
```

```python
male_last_Hb = last_Hb[last_Hb.gender == 'M']
sns.distplot(male_last_Hb['valuenum'], label='male')

female_last_Hb =last_Hb[last_Hb.gender == 'F']
sns.distplot(female_last_Hb['valuenum'],label='female')

# Plot formatting
plt.legend(prop={'size': 12})
plt.title('Last hemoglobin before blood transfusion')
plt.xlabel('valuenum')
plt.ylabel('Density')
```

```python
ax = sns.boxplot(x="gender", y="valuenum", data=last_Hb)
```
